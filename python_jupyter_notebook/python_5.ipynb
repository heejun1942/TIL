```python
def get_html(url):
    html = ''
    res = requests.get(url)
    if res.status_code == 200:
        res.encoding = 'None'
        html = res.text
    return html
```

```python
result = get_html('http://ggoreb.com/python/html/data1.html')
result
```

```python
s_idx=0
e_idx=0
while True:
    s_idx = result.find('<td>',s_idx)
    if s_idx== -1: break
    e_idx = result.find('</td>')
    print(result[s_idx:e_idx])
    s_idx=e_idx
```
